```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
with open('eval_diverse_result.json', encoding='utf-8') as f:
    data = json.load(f)
with open('eval_diverse_result_simclr.json', encoding='utf-8') as f:
    data_clr = json.load(f)
```

```python
x = np.arange(40)
x_data = data['rouge1']
x_data_clr = data_clr['rouge1']
plt.figure(figsize=(10,5))
plt.title("rouge1")
plt.xlabel("return size")
# plt.xticks(rotation=45)
plt.ylabel("rouge1")
plt.plot(x, x_data, label='groundtruth')
plt.plot(x, x_data_clr, label='baseline')
plt.legend()
x_ticks = np.arange(41)[::5]
plt.xticks(x_ticks, x_ticks+1)
# plt.plot_date(x_date,data['close'],'-',label="收盘价")
# plt.legend()
plt.grid()
```
