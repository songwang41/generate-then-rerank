```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
with open('lab_statis/train_std_distribution.json', encoding='utf-8') as f:
    train_data = json.load(f)

with open('lab_statis/test_std_distribution.json', encoding='utf-8') as f:
    test_data = json.load(f)

# print(len(train_data['rouge1_std']))
# print(len(train_data['rouge2_std']))
# print(len(train_data['rougeLsum_std']))
# print(len(train_data['similarity_std']))

# print(len(test_data['rouge1_std']))
# print(len(test_data['rouge2_std']))
# print(len(test_data['rougeLsum_std']))
# print(len(test_data['similarity_std']))
```

```python
fig,ax = plt.subplots(2,2,figsize=(12,12))
ax[0,0].hist(train_data['rouge1_std'], bins=40, label='train', alpha=0.5)
ax[0,0].hist(test_data['rouge1_std'], bins=40, label='test', alpha=0.5)
ax[0,0].set_xlabel('rouge1_std')
ax[0,0].legend()

ax[0,1].hist(train_data['rouge2_std'], bins=40, label='train', alpha=0.5)
ax[0,1].hist(test_data['rouge2_std'], bins=40, label='test', alpha=0.5)
ax[0,1].set_xlabel('rouge2_std')
ax[0,1].legend()

ax[1,0].hist(train_data['rougeLsum_std'], bins=40, label='train', alpha=0.5)
ax[1,0].hist(test_data['rougeLsum_std'], bins=40, label='test', alpha=0.5)
ax[1,0].set_xlabel('rougeLsum_std')
ax[1,0].legend()

ax[1,1].hist(train_data['similarity_std'], bins=40, label='train', alpha=0.5)
ax[1,1].hist(test_data['similarity_std'], bins=40, label='test', alpha=0.5)
ax[1,1].set_xlabel('similarity_std')
ax[1,1].legend()

plt.show()

```

```python
with open('lab_statis/train_rouge_distribution.json', encoding='utf-8') as f:
    train_data = json.load(f)

with open('lab_statis/test_rouge_distribution.json', encoding='utf-8') as f:
    test_data = json.load(f)
```

```python
fig,ax = plt.subplots(2,2,figsize=(12,12))
ax[0,0].hist(np.array(train_data['rouge1'])*100, bins=40, label='train', alpha=0.5)
ax[0,0].hist(np.array(test_data['rouge1'])*100, bins=40, label='test', alpha=0.5)
ax[0,0].set_xlabel('rouge1')
ax[0,0].legend()

ax[0,1].hist(np.array(train_data['rouge2'])*100, bins=40, label='train', alpha=0.5)
ax[0,1].hist(np.array(test_data['rouge2'])*100, bins=40, label='test', alpha=0.5)
ax[0,1].set_xlabel('rouge2')
ax[0,1].legend()

ax[1,0].hist(np.array(train_data['rougeLsum'])*100, bins=40, label='train', alpha=0.5)
ax[1,0].hist(np.array(test_data['rougeLsum'])*100, bins=40, label='test', alpha=0.5)
ax[1,0].set_xlabel('rougeLsum')
ax[1,0].legend()

ax[1,1].hist(train_data['similarity'], bins=40, label='train', alpha=0.5)
ax[1,1].hist(test_data['similarity'], bins=40, label='test', alpha=0.5)
ax[1,1].set_xlabel('similarity')
ax[1,1].legend()

plt.show()
```
